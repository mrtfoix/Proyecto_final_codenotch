```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fechas = [
    '1/3',
    '14/7',
    '25/11'
]
```

```python
dias_por_mes = [
    31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31
]
```

```python
def dia_del_a√±o(fecha):
    dia, mes = fecha.split('/')
    dia = int(dia)
    mes = int(mes)
    return dia + sum(dias_por_mes[:mes-1])
```

```python
fechas = []
for i, mes in enumerate(dias_por_mes):
    for dia in range(dias_por_mes[i]):
        dia = dia + 1
        mes = i + 1
        fechas.append(f'{dia}/{mes}')
```

```python
fechas
```

```python
for fecha in fechas:
    angulo = 2*np.pi * dia_del_a√±o(fecha) / 365
    x = np.cos(angulo)
    y = np.sin(angulo)
    plt.scatter(x, y, c = 'b')
```
